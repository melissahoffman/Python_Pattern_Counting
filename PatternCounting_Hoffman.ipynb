### Functions

Count the number of each base across all sequences

```python
counter = [0,0,0,0]
for line_num,line in enumerate(fastq):  #number all of the lines in fastq
    if line_num%4 == 1:  #specifying the loop to target every 4th line
        for base in line:
            if base == 'A':
                counter[0] = counter[0]+1   #if counter finds an A, add one to the total count of A's in the seq
            elif base == 'C':
                counter[1] = counter[1]+1   #use elif for "else if" statement
            elif base == 'G':
                counter[2] = counter[2]+1
            elif base == 'T':
                counter[3] = counter[3]+1
print(counter)
```

Count the number of each pair of bases (described as kmers of size 2).

```python
def count_kmers2():   #starting a function
    counter ={'AA':0, 'GT':0, 'TA':0, 'CG':0}    #specifying combinations of pairs
    for line_num, line in enumerate(fastq[0:40]):    #narrowing the amount of data the function is processing
        if line_num%4==1:    #looking at every 4th line
            line=line.rstrip()   #returns copy of the string
            for base_number, base in enumerate(line):   #counting the number of bases in each line that was called
                if base_number != (len(line)-1):   #don't count a base if length of line...
                    kmer = line[base_number]+line[base_number +1]   #defining kmer so it adds when recognized
                    if kmer in counter:    #if kmer is any of the combinations in line 2 of this function
                        counter[kmer]+=1
    print(counter)
```

```python
count_kmers2()
```

Count the number of each pair of bases, without  assuming you know in advance the possible pairs (i.e. start with an empty dictionary).

```python
def count_BP_unknown(): 
    counter ={}    #using an empty dictionary, not assuming possible pairs
    for line_num, line in enumerate(fastq[0:40]):  
        if line_num%4==1:     #looking at every 4th line
            line=line.rstrip()   
            for base_number, base in enumerate(line): 
                if base_number != (len(line)-1): 
                    kmer = line[base_number]+line[base_number +1]   #locates first position in the kmer, and then the second with the "+1"
                    if kmer in counter:    #if the kmer is recognized in the dictionary then count it
                        counter[kmer]+=1     #since the dictionary is empty, need an "else"
                    else: counter[kmer]=1      #if the kmer isn't in the dictionary, then define as a kmer
    print(counter)
```



Alter the counting function to allow for counting kmers of any size.

```python
def count_AnyKmer(): 
    counter ={}  
    for line_num, line in enumerate(fastq[0:2]):  
        if line_num%4==1:   
            line=line.rstrip()   
            for base_number, base in enumerate(line): 
                if base_number != (len(line)-1): 
                    for end_number, end in enumerate(line[base_number:len(line)]): 
                        #enumerate to the end of the line
                        kmer = line[base_number:(base_number + end_number)] 
                        #start counting with base number as the first position in the kmer
                        #the ending of the kmer be after the first position
                        #the loop let's the base number position continuously change, as well as the end number
                        if kmer in counter:   
                            counter[kmer]+=1  
                        else: counter[kmer]=1  
    print(counter)
```
